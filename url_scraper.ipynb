```python
import requests

from bs4 import BeautifulSoup
from selenium import webdriver

import pandas as pd
```

```python
def scrape_yelp_business_urls(city, state, industry):
    base_url = 'https://www.yelp.com/search'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    start = 0
    business_urls = []

    while True:
        # Construct the parameters with start index
        params = {
            'find_desc': industry,
            'find_loc': f"{city}, {state}",
            'sortby': 'recommended',
            'start': start
        }

        # Send a request to the URL with headers
        response = requests.get(base_url, params=params, headers=headers)

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all business name elements
        business_elements = soup.find_all('h3', class_='y-css-hcgwj4')

        # Check if there are any more businesses
        if not business_elements:
            break

        # Loop through each business element to extract the URL
        for business_element in business_elements:
            # Find the anchor tag within the h3 element
            anchor_tag = business_element.find('a', class_='y-css-12ly5yx')
            if anchor_tag:
                # Extract the href attribute (URL) from the anchor tag (has the URL extension stored in the HTML)
                business_url = 'https://www.yelp.com' + anchor_tag.get('href')
                # Check if URL is not an adredir link (these are ads, not sure if we want these).
                if 'adredir' not in business_url:
                    business_urls.append(business_url)

        # Increase start index for the next page
        start += 10

    # Print the list of business URLs
    for idx, url in enumerate(business_urls, start=1):
        print(f"Business {idx}: {url}")

# Define City, State and Industry
city = 'Philadelphia'
state = 'PA'
industry = 'Restaurants'

# Call the function
scrape_yelp_business_urls(city, state, industry)
```
